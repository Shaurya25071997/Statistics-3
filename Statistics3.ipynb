## Q1. Write the Gaussian Distribution empirical formula?

Gaussian distribution emperical formula basicall starts from a emperical rule which is 68-95-99.7 % rule and this distribution is having symmetric distribution

Here we can see the formula Is:- 
    
1. P [ μ – σ <= X <= μ + σ ] ≈ 68 %

This formula implies 68% of the data points lies between  μ – σ to  μ + σ

2. P [ μ – 2σ <= X <= μ + 2σ ] ≈ 95 %

This formula implies 95% of the data points lies between  μ – 2σ to  μ + 2σ

3.  P [ μ – 3σ <= X <= μ + 3σ ] ≈ 99.7 %

This formula implies 95% of the data points lies between  μ – 3σ to  μ + 3σ\

## Q.3 What is an outlier, exactly?

An outlier is an observation that lies an abnormal distance from other values in a dataset Outliers hold useful informtaion of data and for better understanding 

If we wnat to find outlier we can use 5 number summary as well as use these two formula given below:- 

1. Outlier < Q1 - 1.5(IQR)
2. Outlier > Q3 + 1.5(IQR)

Exmaple to find Outliers:-
    
step:-1 data={1,2,2,2,3,3,4,5,5,,5,6,,6,6,6,7,7,8,8,9,27}
step:2 Lower fence = Q1 - 1.5(IQR)   #IQR denotes Interquartile range
       Higher fence = Q3 + 1.5(IQR)  
        
Q1(25%) = percentile*100/n+1
        = 25*100/19+1  = 5th Index = 3
    
Q3(75%) = percentile*100/n+1
        = 75*100/19+1  = 10th Index  = 7
    
IQR = Q3 - Q1
IQR = 7  - 3
IQR = 4

Now we will use this in lower fence and higher fence

Lower fence = Q1-1.5(IQR) => 3 -1.5(4) => -3
Higher fence = Q3+1.5(IQR) => 7 +1.5(4) => 13

Lowerfence <---------------> Higher fence
-3 <--------------> 13

so here we can see 27 is a outlier 

data = {1,2,2,2,3,3,3,4,5,5,5,6,6,6,6,7,8,8,9}

so here you can see 27 is been removed because it is an outlier
